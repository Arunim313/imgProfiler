```python
import cv2
import numpy as np

# Read the image
image = cv2.imread(r"D:\Projects\image_profiler\test_use\traffic.jpg")

# Create a sharpening kernel
kernel = np.array([[-1,-1,-1], 
                   [-1, 9,-1],
                   [-1,-1,-1]])

# Apply the kernel to the image
sharpened = cv2.filter2D(image, -1, kernel)

# Save or display the result
cv2.imwrite('CV2sharpened_image.jpg', sharpened)
```
